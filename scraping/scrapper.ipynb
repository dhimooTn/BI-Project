```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from HelloWorkScraper import HelloWorkScraper
```

```python
# -------------------------------------------------
# CHROME OPTIONS (ANTI-BAN BASIQUE)
# -------------------------------------------------
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option("useAutomationExtension", False)
```

```python
service = Service()  # utilise chromedriver auto

driver = webdriver.Chrome(service=service, options=chrome_options)

# -------------------------------------------------
# SCRAPER
# -------------------------------------------------
base_url = "https://www.hellowork.com/fr-fr/emploi/recherche.html?page={}"

scraper = HelloWorkScraper(
    driver=driver,
    base_url=base_url,
    wait_time=10
)
scraper.scrape_all_pages(start_page=1, end_page=400)

# -------------------------------------------------
# RESULT
# -------------------------------------------------
```

```python
df = scraper.to_dataframe()
print(df.head())
print(df.info())
print(f"\nâœ… Total offres scrapÃ©es : {len(df)}")
```

```python
# -------------------------------------------------
# CLEAN EXIT
# -------------------------------------------------
driver.quit()
```
